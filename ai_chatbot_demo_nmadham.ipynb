# PreRequisite: Import pandas and data file uploaded into a google folder

```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

# Step1: Read data file into Dataframe

```python
file_path='/content/drive/My Drive/Chatbot_Demo/10K_financials_chatbot.csv'
df = pd.read_csv(file_path)
display(df)
```

# Step2: Convert a set of String columns to Float

```python
columns_to_int = ['Total Revenue','Net Income','Total Assets','Total Liabilities','Cash Flow']
# Remove commas from specified columns
df[columns_to_int] = df[columns_to_int].apply(lambda x: x.str.replace(',', ''))
```

```python
# Convert specified columns to int
df_new = df.copy()
df_new[columns_to_int] = df_new[columns_to_int].astype(float)

# Print the new DataFrame
display(df_new)
```

# Scatter plot on the above derived Metrics

```python
#Using dataframe df_new: scatter

import altair as alt
chart = alt.Chart(df_new).mark_circle().encode(
    x = alt.X('Total Revenue', title='Total Revenue (in billions)'),
    y = alt.Y('Net Income', title='Net Income (in billions)'),
    color = alt.Color('Year', legend=alt.Legend(title='Year'))
).properties(width=600)
chart
```

# Step3: Group by Year and calculate sum of Net Income

```python
# Group by Year and calculate sum of Net Income
total_income_by_year = df_new.groupby('Year')['Net Income'].sum()

# Display the results
print("\nTotal Income by Year:\n")
print(total_income_by_year)
```

# Step4: Group by Company and calculate mean of Total Revenue

```python
# Group by Company and calculate mean of Total Revenue
average_revenue_by_company = df_new.groupby('Company')['Total Revenue'].mean()

# Display the results
print("Average Revenue by Company:\n")
print(average_revenue_by_company)
```
